```python
import pandas as pd
import numpy as np
```

```python
ds1=pd.read_csv("Retail_datasets_P2/user_demographics.csv")
```

```python
ds1.shape
```

```python
ds2=pd.read_csv("Retail_datasets_P2/User_product_purchase_details_p2.csv")
```

```python
ds2.shape
```

```python
ds=pd.merge(ds1,ds2,on='User_ID')
```

```python
ds
```

```python
ds.isna().sum()
```

```python
ds.info()
```

```python
ds['Product_Category_2']=ds['Product_Category_2'].fillna(0.0)
```

```python
ds['Product_Category_3']=ds['Product_Category_3'].fillna(0.0)
```

```python
ds['Product_Category_1']=ds['Product_Category_1'].fillna(0.0)
```

```python
ds.info()
```

```python
ds[['Product_Category_2','Product_Category_3']]=ds[['Product_Category_2','Product_Category_3']].astype(int)
```

```python
ds.info()
```

```python
ds.drop_duplicates()
```

```python
# split the range of age
ds['start']=ds['Age'].str.split('-').str[0]
ds['end']=ds['Age'].str.split('-').str[1]
# + present so replace it!
ds['start']=ds['start'].str.replace('+','',regex=False)
# for null values in end age replace it with start of range
ds['end']=ds['end'].fillna(ds['start'])
# convert the type to int
ds['start']=ds['start'].astype(int)
ds['end']=ds['end'].astype(int)
# find the median of age
ds['Age']=(ds['start']+ds['end'])/2
#drop the temporary column
ds.drop(['start','end'],axis=1,inplace=True)
```

```python
ds['Age']=ds['Age'].astype(int)
```

```python
ds.info()
```

```python
new_ds=ds[['User_ID','Gender','Age','Occupation','Product_ID','City_Category','Stay_In_Current_City_Years','Marital_Status','Product_Category_1','Product_Category_2','Product_Category_3','Purchase']].groupby('User_ID').sum()
```

```python
new_ds
```

```python
new_ds=new_ds.reset_index()
```

```python
new_ds.drop(['Age','Gender','City_Category','Stay_In_Current_City_Years','Marital_Status','Occupation'],axis=1,inplace=True)
```

```python
new_ds
```

```python
remove_df=ds[['User_ID','Age','Gender','City_Category','Stay_In_Current_City_Years','Marital_Status','Occupation']]
```

```python
remove_df=remove_df.drop_duplicates()
```

```python
type(new_ds)
```

```python
unique_ds=pd.merge(remove_df,new_ds,on='User_ID')
```

```python
unique_ds
```

```python
unique_ds.info()
```

```python
unique_ds['Stay_In_Current_City_Years']=unique_ds['Stay_In_Current_City_Years'].astype(str)
```

```python
unique_ds['Stay_In_Current_City_Years']=unique_ds['Stay_In_Current_City_Years'].str.replace('+','')
```
