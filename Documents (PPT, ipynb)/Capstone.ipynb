```python
import numpy as np 
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,accuracy_score,r2_score
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
df=pd.read_csv('user_demographics.csv')
```

```python
df
```

```python
# Splitting the age range into int(age)
df['start'] = df['Age'].str.split('-').str[0]
df['end'] = df['Age'].str.split('-').str[1]

# '+' is present so replace
df['start'] = df['start'].str.replace('+', '', regex=False)

# For null values in the end of range, replace with start
df['end'] = df['end'].fillna(df['start'])

# Convert to int datatype
df['start'] = df['start'].astype(int)
df['end'] = df['end'].astype(int)

# Finding median
df['Age'] = (df['start'] + df['end']) / 2

# Drop the temporary columns
df.drop(['start', 'end'], axis=1, inplace=True)

# Ensure Age is an integer
df["Age"] = df['Age'].astype(int)
```
